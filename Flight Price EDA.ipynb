# NOW LET CHECK HOW MANY FLIGHTS OF DIFFERENT AIRLINES ARE PROVIDING SERVICE IN THE SAME ROUTE:

```python
new_train['Route'].value_counts()
# BY THIS RESULT IT'S CLEAR THAT THERE ARE TOP FIVE ROUTE IN WHICH  THE AIRLINES HAS PROVIDED MOST SERVICE TO THEIR 
# COMMUTERS DELHI→BOMBAY→COACHIN THEN  BLR→DEL,CCU→BOM→BLR AND SO ON..
```

 1.BY THIS WE CAN CHECK THAT WHAT IS THE FARE PRICE OF DIFFERENT AIRLINES FOR THE SAME ROUTE WITH 
 ALL SAME SERVICES AND SAME NUMBER OF STOPS THERE MAY BE  SLIGHTLY DIFFERENCE IN THE TIME TAKEN THROUGHOUT THE JOURNEY.

```python
Route1=new_train[new_train['Route']=='DEL → BOM → COK']
Route1    
```

```python
Route1['Airline'].value_counts()
```

```python
airline_prices = Route1.groupby('Airline')['Price'].mean().reset_index()
airline_prices.sort_values(by="Price")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(airline_prices['Airline'], airline_prices['Price'])
plt.xlabel('Airline',color='green',fontsize=20,weight='bold')
plt.ylabel('Price',weight='bold',fontsize=20)
plt.xticks(rotation=30,weight='bold')
plt.yticks(rotation=30,color='red')
plt.title('Average Prices of Airlines for the same route [DEL→BOM→COK]',fontsize=18)
plt.show()
```

BY ABOVE VISUALISATION THE GO AIR AIRLINE HAS LOW TICKETS PRICE AND THE JET AIRWAYS ARILINE HAS HIGHEST TICKETS PRICE FOR 
THE ROUTE OF [DEL→ BOM → COK]

```python
airline_time = Route1.groupby('Airline')['Total_time(in M)'].mean().reset_index()
airline_time.sort_values(by="Total_time(in M)")
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Airline',y='Total_time(in M)',data=airline_time)
plt.xticks(rotation=30)
plt.xlabel('Airline',weight='bold',fontsize=16)
plt.ylabel('Total_time(in M)',weight='bold',fontsize=16)
plt.title('Time taken by different Airlines for Route1[DEL → BOM → COK ]',weight='bold')
```

THE TIME TAKEN BY THE GOAIR AIRLINE IS LESS THAN THE OTHERS AIRLINES FOR THE SAME ROUTE [DEL → BOM → COK] WHILE JET AIRWAYS
AIRLINE TAKE MORE TIME FOR SIMILAR ROUTE .
